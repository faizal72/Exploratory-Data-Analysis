# 4) Data Cleaning

```python
data.isnull().sum()
```

```python
# handling nan values in name columns 
data['name'] = data.name.fillna('xyz')
data['name'].isnull().sum()
```

```python
# handling duplicate values in host name columns
data['host_name'] = data.host_name.fillna('xyz')
data['host_name'].isnull().sum()
```

```python
# fill mean value 
mean_r_months=data.reviews_per_month.mean()
data['reviews_per_month'] = data['reviews_per_month'].fillna(mean_r_months)
data['reviews_per_month'].isnull().sum()
```

```python
data.isnull().sum()
```

```python
# cheking duplicate row
data.duplicated().sum()
```

# 5) Data Analysis

Univariate Analysis


```python
# identifying outliers in price column

df = data[data['price'] <2000]
sns.boxplot(data= df, x = 'price')
```

```python
# price Distribution
plt.figure(figsize = (8,5))
sns.histplot(data=df, x = 'price', bins= 100)
plt.title('Price Distributions')
plt.ylabel('Frequency')
plt.show()
```

```python
# availability_365 distributions
plt.figure (figsize = (8,5))
sns.histplot(data=df,x = 'availability_365')
plt.title('Availability_365 Distribution')
plt.ylabel('Frequency')
plt.show()
```

```python
df.groupby(by= 'neighbourhood_group')['price'].mean()
```

 Feature Enginerring

```python
df ['price_per_night'] = df['price'] / df['minimum_nights']
```

```python
# avg price per night
df.groupby(by = 'neighbourhood_group')['price_per_night'].mean()
```

```python
# Price dependancy on neighbourhood
sns.barplot(data=df, x = 'neighbourhood_group', y= 'price', hue = 'room_type')
plt.title('Price dependancy on neighbourhood')
plt.show()
```

```python
# number_of_reviews and price relationship
plt.figure (figsize = (8,5))
sns.scatterplot(data = df , x = 'number_of_reviews' , y ='price', hue = 'neighbourhood_group')
plt.title('Locality and Review Dependency')
plt.show()
```
